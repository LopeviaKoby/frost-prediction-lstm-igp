```python
missing_summary = (
    raw_df.isna()
    .sum()
    .sort_values(ascending=False)
    .rename('missing_count')
    .to_frame()
)
missing_summary['missing_pct'] = (missing_summary['missing_count'] / len(raw_df) * 100).round(4)
missing_summary
```
