# EDA v01: Panorama General de los Datos

Este notebook explora la estacionalidad, la variable objetivo contemporanea y el comportamiento general de las series antes de definir el problema de pronostico.


## Enfoque

El objetivo aqui no es agotar cada grafico posible, sino construir intuicion sobre:

- frecuencia de heladas;
- distribucion mensual y horaria;
- relaciones simples entre variables meteorologicas;
- posibles señales de ruido o outliers.


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
while PROJECT_ROOT.name != 'frost_ema_igp' and PROJECT_ROOT != PROJECT_ROOT.parent:
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

PROJECT_ROOT
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from src.data.loaders import load_raw_hourly_dataset

raw_df = load_raw_hourly_dataset()
raw_df['frost_event_current'] = (raw_df['tempsup_min'] <= 0).astype(int)
raw_df[['tempsup_min', 'HR_mean', 'radinf_mean']].describe().T
```

```python
monthly_frost = raw_df.assign(month=raw_df.index.month).groupby('month')['frost_event_current'].mean()
hourly_frost = raw_df.assign(hour=raw_df.index.hour).groupby('hour')['frost_event_current'].mean()

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
monthly_frost.plot(kind='bar', ax=axes[0], title='Tasa de helada por mes')
hourly_frost.plot(kind='bar', ax=axes[1], title='Tasa de helada por hora')
plt.tight_layout()
plt.show()
```

```python
correlation_cols = ['tempsup_mean', 'tempsup_min', 'HR_mean', 'radinf_mean', 'vel_mean', 'press_mean']
raw_df[correlation_cols].corr()
```
